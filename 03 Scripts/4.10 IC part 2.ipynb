# Create masks for dependants and family types

```python
mask = (df_cust_depts['n_dependants'] > 0) & (df_cust_depts['department'] == 'babies')
df_cust_depts['dependent_type'] = df_cust_depts.groupby('user_id')['user_id'].transform(
    lambda x: 'babies in household' if any(mask[x.index]) else 'no babies in household')
```

```python
df_cust_depts.head()
```

```python
mask2 = (df_cust_depts['n_dependants'] > 0) & (df_cust_depts['age_group'] == '18-29')
df_cust_depts['parental_age'] = df_cust_depts.groupby('user_id')['user_id'].transform(
    lambda x: 'young parent' if any(mask2[x.index]) else 'older parent')
```

```python
df_cust_depts.loc[(df_cust_depts['n_dependants'] > 0) & (df_cust_depts['age_group'] == 0), '18-29'] = 'young parent'
```

```python
df_cust_depts['parental_age'].value_counts()
```

```python
df_cust_depts.loc[(df_cust_depts['fam_status'].isin(['divorced/widowed', 'living with parents and siblings', 'single'])) & (df_cust_depts['n_dependants'] == 0), 'family_type'] = 'single no dependents'
df_cust_depts.loc[(df_cust_depts['fam_status'].isin(['divorced/widowed', 'living with parents and siblings', 'single'])) & (df_cust_depts['n_dependants'] > 0), 'family_type'] = 'Single with dependents'
df_cust_depts.loc[(df_cust_depts['fam_status'] == 'married') & (df_cust_depts['n_dependants'] == 1), 'family_type'] = 'married no dependents'
df_cust_depts.loc[(df_cust_depts['fam_status'] == 'married') & (df_cust_depts['n_dependants'] > 1), 'family_type'] = 'married with dependents'
```
