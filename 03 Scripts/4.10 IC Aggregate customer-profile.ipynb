# Create groups for each aggregate

```python
age_agg = df.groupby('age_group').agg({'order_id': 'count', 'days_since_last_order': ['min', 'mean', 'max'], 'prices': ['min', 'mean', 'max']})
```

```python
income_agg = df.groupby('income_group').agg({'order_id': 'count', 'days_since_last_order': ['min', 'mean', 'max'], 'prices': ['min', 'mean', 'max']})
```

```python
family_type_agg = df.groupby('family_type').agg({'order_id': 'count', 'days_since_last_order': ['min', 'mean', 'max'], 'prices': ['min', 'mean', 'max']})
```

```python
regional_agg = df.groupby('region').agg({'order_id': 'count', 'days_since_last_order': ['min', 'mean', 'max'], 'prices': ['min', 'mean', 'max']})
```

```python
dependent_type_agg = df.groupby('dependent_type').agg({'order_id': 'count', 'days_since_last_order': ['min', 'mean', 'max'], 'prices': ['min', 'mean', 'max']})
```

```python
parental_age_agg =df.groupby('parental_age').agg({'order_id': 'count', 'days_since_last_order': ['min', 'mean', 'max'], 'prices': ['min', 'mean', 'max']})
```

```python
loyalty_agg = df.groupby('loyalty_flag').agg({'order_id': 'count', 'days_since_last_order': ['min', 'mean', 'max'], 'prices': ['min', 'mean', 'max']})
```

```python
fams_agg = df.groupby('fam_status').agg({'order_id': 'count', 'days_since_last_order': ['min', 'mean', 'max'], 'prices': ['min', 'mean', 'max']})
```
