## The difference is we can see the entire dataframe so we can see all the departments, in the sample we could only see some of them

```python
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_num_of_customer'].transform(np.max)
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

## the mean for loyal customers and regular customers is less than the mean for new customers. Also the max prices is the same for every group

```python
ords_prods_merge['purchase_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.loc[ords_prods_merge['purchase_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['purchase_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
ords_prods_merge.head()
```

```python
ords_prods_merge['median_purchase'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge.loc[ords_prods_merge['median_purchase'] > 20, 'order_fr_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_purchase'] > 10 & (ords_prods_merge['median_purchase'] <= 20), 'order_fr_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_purchase'] <= 10, 'order_fr_flag'] = 'Frequent customer'
```
