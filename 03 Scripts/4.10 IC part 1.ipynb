# Create customer profiles

```python
df_active.loc[df_active['age'] <= 29, 'age_group'] = '18-29'
df_active.loc[(df_active['age'] > 29) & (df_active['age'] < 40), 'age_group'] = '30-39'
df_active.loc[(df_active['age'] > 39) & (df_active['age'] < 50), 'age_group'] = '40-49'
df_active.loc[(df_active['age'] > 49) & (df_active['age'] < 60), 'age_group'] = '50-59'
df_active.loc[(df_active['age'] > 59) & (df_active['age'] < 70), 'age_group'] = '60-69'
df_active.loc[df_active['age'] > 69, 'age_group'] = '70 and over'
```

```python
df_active['age_group'].value_counts(dropna = False)
```

```python
df_active.loc[df_active['income'] <= 75000, 'income_group'] = 'Working'
df_active.loc[(df_active['income'] > 75000) & (df_active['income'] < 120001), 'income_group'] = 'Middle'
df_active.loc[(df_active['income'] > 120000) & (df_active['income'] < 400001), 'income_group'] = 'Upper'
df_active.loc[df_active['income'] > 400000, 'income_group'] = 'High Wealth'
```

```python
df_active['income_group'].value_counts(dropna = False)
```
